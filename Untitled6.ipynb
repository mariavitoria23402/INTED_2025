```python
#14
soma_idades = 0
contador_idades = 0

print("Digite as idades (digite um n√∫mero negativo para encerrar):")

idade = 0

while idade >= 0:
    idade = int(input("Idade: "))

    if idade >= 0:
        soma_idades += idade
        contador_idades += 1

if contador_idades > 0:
    media = soma_idades / contador_idades
    print(f"Total de idades v√°lidas: {contador_idades}")
    print(f"A m√©dia das idades digitadas √©: **{media:.2f}**")
else:
    print("Nenhuma idade v√°lida foi digitada.")
```

```python
#23
positivos = 0
negativos = 0
zeros = 0
TOTAL_NUMEROS = 10

print(f"Digite {TOTAL_NUMEROS} n√∫meros:")

for i in range(TOTAL_NUMEROS):
    numero = float(input(f"N√∫mero #{i + 1}: "))

    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1

print("Resultado da An√°lise")
print(f"N√∫meros Positivos: {positivos}")
print(f"N√∫meros Negativos: {negativos}")
print(f"N√∫meros Iguais a Zero: {zeros}")
```

```python
#24
nomes_filtrados = []
TOTAL_NOMES = 6

print(f"Digite {TOTAL_NOMES} nomes:")

for i in range(TOTAL_NOMES):
    nome = input(f"Nome #{i + 1}: ").strip().capitalize()
    if nome.startswith('A') or nome.startswith('C'):
        nomes_filtrados.append(nome)

print("Nomes que come√ßam com 'A' ou 'C'")
if nomes_filtrados:
    for nome in nomes_filtrados:
        print(f" {nome}")
else:
    print("Nenhum nome digitado come√ßa com 'A' ou 'C'.")
```

```python
#27
SENHA_CORRETA = "python123"
MAX_TENTATIVAS = 3

print(f"Sistema de Login (M√°ximo de {MAX_TENTATIVAS} tentativas):")

for tentativa in range(1, MAX_TENTATIVAS + 1):
    senha_digitada = input(f"Tentativa {tentativa}/{MAX_TENTATIVAS} - Digite a senha: ")

    if senha_digitada == SENHA_CORRETA:
        print("Sucesso! Login efetuado com sucesso!")
        break
    else:
        if tentativa < MAX_TENTATIVAS:
            print("Senha incorreta. Tente novamente.")
        else:
            print(" Login bloqueado. Voc√™ excedeu o n√∫mero m√°ximo de tentativas.")
```

```python
#28
soma_positivos = 0
TOTAL_NUMEROS = 10

print(f"Digite {TOTAL_NUMEROS} n√∫meros para somar apenas os positivos:")

for i in range(TOTAL_NUMEROS):
    numero = float(input(f"N√∫mero #{i + 1}: "))

    if numero < 0:
        print("(N√∫mero negativo, ignorado.)")
        continue

    soma_positivos += numero

print(f"A soma dos n√∫meros positivos digitados √©: **{soma_positivos:.2f}**")
```

```python
#29
soma = 0
limite = 200
numero_atual = 1

print(f"Acumulando n√∫meros at√© a soma ultrapassar {limite}...")

while True:
    soma += numero_atual

    print(f"Somando {numero_atual} -> Soma atual: {soma}")

    if soma > limite:
        print(f"Limite de {limite} ultrapassado!")
        break

    numero_atual += 4

print(f"A soma final √©: **{soma}**")
```
