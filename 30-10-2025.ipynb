```python
#8 Peça um número de 1 a 7 e informe o dia da semana correspondente. Crie um algoritmo em Python que associe o número ao dia correto.
num = int(input('digite um número de 1 a 7 informando o dia da semana correspondente:'))

if num == 1:
  print("Domingo")
elif num == 2:
  print("Segunda-feira")
elif num == 3:
  print("Terça-feira")
elif num == 4:
  print("Quarta-feira")
elif num == 5:
  print("Quinta-feira")
elif num == 6:
  print("Sexta-feira")
elif num == 7:
  print("Sábado")
else:
  print("Número inválido. Por favor, digite um número de 1 a 7.")
```

```python
#9 Peça um ano e informe se ele é bissexto. Crie um algoritmo em Python que verifique se o ano é divisível por 4 e não por 100, exceto se também for divisível por 400.
num = int(input('digite um ano e informe se é bissexto:'))
if (num % 4 == 0 and num % 100 != 0) or (num % 400 == 0):
  print(f"{num} é um ano bissexto")
else:
  print(f"{num} não é um ano bissexto")
```

```python
#10 Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.
horas_trabalhadas = int(input('digite a quantidade de horas trabalhadas:'))
valor_por_hora = float(input('digite o valor por hora:'))

if horas_trabalhadas <= 40:
  salario = horas_trabalhadas * valor_por_hora
  print(f"O salário é: R$ {salario:.2f}")
else:
  horas_normais = 40
  horas_extras = horas_trabalhadas - horas_normais
  salario_normal = horas_normais * valor_por_hora
  salario_extra = horas_extras * valor_por_hora * 1.5
  salario_total = salario_normal + salario_extra
  print(f"O salário é: R$ {salario_total:.2f}")
```
